# Traits to CSV

```python
import json
import pandas as pd
```

Read json file

```python
with open('../data/traits/traitsArray.json',encoding="utf8") as json_file:
    data = json.load(json_file)
```

Create a Dataframe

```python
df = pd.DataFrame()
for i in range(10000):
    attributes = data[i]['attributes']
    dft = pd.json_normalize(attributes).drop('_id', axis=1, errors='ignore').transpose()
    dft.columns = dft.loc['trait_type']
    dft = dft.drop('trait_type').reset_index(drop=True)
    dft.columns.name = None
    df = pd.concat([df, dft], ignore_index=True, copy=False)
df
```

```python
df.columns
```

Are there duplicated values?

```python
df[df.duplicated(keep=False)].sort_values(by=list(df.columns),axis=0)
```

Dataframe to CSV

```python
df.to_csv('../data/traits/traits.csv', index=False)
```
